```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

sm_urls = ["https://www.smsupermalls.com/mall-directory/metro-manila/",
           "https://www.smsupermalls.com/mall-directory/north-luzon/",
           "https://www.smsupermalls.com/mall-directory/south-luzon/",
           "https://www.smsupermalls.com/mall-directory/visayas/",
           "https://www.smsupermalls.com/mall-directory/mindanao/"]

websiteURL = "https://www.smsupermalls.com/mall-directory/metro-manila/"
```

```python
image_link = []
mall_link = []
mall_name = []
for url in sm_urls:
    website = (requests.get(url).text) 
    soup = BeautifulSoup(website, 'lxml')
    mallDiv = soup.find("div", {"id": "mall-locator-directory", "class": "col-xs-12 col-sm-9 col-md-9 col-lg-9"})
    image_link.extend(list(map(lambda a_tag: a_tag.find('img')['src'], mallDiv.find_all('a'))))
    mall_name.extend(list(map(lambda a_tag: a_tag.find("div", {"class": 'padding-5 clearfix text-black text-center'}).get_text(), mallDiv.find_all('a'))))
    mall_link.extend(list(map(lambda a_tag: a_tag['href'], mallDiv.find_all('a'))))
```

```python
len(image_link)
```

```python
import random
import time
mall_address = []
mall_contact_no = []
for link in mall_link:
    #link = "https://www.smsupermalls.com/mall-directory/sm-center-angono/information"
    website = (requests.get(link).text) 
    soup = BeautifulSoup(website, 'lxml')
    mall_info = soup.find("div", {"id": "mall-info"})
    mall_details = mall_info.find("div", {"class": "col-sm-7"}).find_all('p')
    mall_address.append(mall_details[0].get_text().strip())
    mall_contact_no.append(mall_details[1].get_text().strip())
    x = random.randint(1,10)
    print(mall_address)
    print(mall_contact_no)
    print(f'sleep for {x} seconds')
    time.sleep(random.randint(1,5))
#print(mall_info)
```

```python
sm_malls_df = pd.DataFrame({'mall_name': mall_name,
                        'mall_address': mall_address,
                        'mall_contact_no': mall_contact_no,
                        'image_link': image_link,
                        'mall_link': mall_link})
```

```python
sm_malls_df
```

```python
import os
directory = 'data'
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
path = os.path.join(directory, "sm.csv")
sm_malls_df.to_csv(path, encoding='utf-8', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import sys

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("chromedriver")
driver.get("https://www.ayalamalls.com/")
#get_Our_Malls = driver.find_element_by_xpath("/html/body/div/div/div/ul/li[4]")
get_Our_Malls = driver.find_element(By.XPATH, "//*[@role='menuitem' and contains(text(),'Our Malls')]")
```

```python
mall_links = driver.find_elements(By.XPATH, "//*[@class='our-malls-list-container']/ul/li")
```

```python
i = 1
for mall in mall_links:
    print(i, mall.text)
    i = i + 1
```

```python
#driver.find_element(By.XPATH, "//*[@class='section-list']/*[@class='section'][3]//*[@class='section-column details' and contains(text(),'+')]").text
driver.find_element(By.XPATH, "//*[@class='section-list']/*[@class='section'][3]//*[@class='section-contents row']/*[@class='section-column details' and contains(text(),'facebook')]").text
```

```python
mall_name = []
image_link = []
logo_link = []
fb_link = []
mall_contact_no = []
mall_address = []
for mall in mall_links:
    #print(x.get_attribute("class"))
    if mall.text != 'Glorietta':
        driver.execute_script("arguments[0].setAttribute('class','active')", mall)
        mall.click()
    print(mall.text)
    mall_name.append(mall.text)
    try:
        #temp = driver.find_element(By.XPATH, "//*[@class='our-malls-mall-details']//img").get_attribute("src")
        image_link.append(driver.find_element(By.XPATH, "//*[@class='our-malls-mall-details']//img").get_attribute("src")) # gets the first image
        #print(temp)
    except:
        image_link.append(None)
    try:
        logo_link.append(driver.find_element(By.XPATH, "//*[@class='mall-logo']/img").get_attribute("src"))
    except:
        logo_link.append(None)
    try:
        mall_address.append(driver.find_element(By.XPATH, "//*[@class='section-list']//*[@class='section-column']").text)
    except:
        mall_address.append(None)
    try:
        #mall_contact_no.append(driver.find_element(By.XPATH, "//*[@class='section-list']/*[@class='section'][3]//*[@class='section-column details']").text)
        mall_contact_no.append(driver.find_element(By.XPATH, "//*[@class='section-list']/*[@class='section'][3]//*[@class='section-column details' and contains(text(),'+')]").text)
    except:
        mall_contact_no.append(None)
    try:
        fb_link.append(driver.find_element(By.XPATH, "//*[@class='section-list']/*[@class='section'][3]//*[@class='section-contents row']/*[@class='section-column details' and contains(text(),'facebook')]").text)
        #print("with fb link")
    except:
        fb_link.append(None)
        #print("no fb link")
    time.sleep(2)
    
    
        
    
```

Scraping Robinsons' malls

```python
url  = "http://www.robinsonsmalls.com/list-of-malls/"
```

```python
driver = webdriver.Chrome("chromedriver")
driver.get(url)
```

```python
website = (requests.get(url).text) 
soup = BeautifulSoup(website, 'lxml')
```

```python
locations = soup.find("ul", {"class": "locations"})
```

```python
locations = locations.find_all("li", {"class": lambda x: x and x.startswith("location-list")})
```

```python
mall_name = []
image_link = []
fb_link = []
mall_link = []
mall_contact_no = []
mall_address = []
i = 0
for x in locations:
    print(x.find("span", {"class": "mall-title"}))
    mall_name.append(x.find("span", {"class": "mall-title"}).text)
    mall_address.append(x.find("div", {"class": "location-address"}).text)
    print(x.find("div", {"class": "location-address"}).text)

    code = x.find("a")['onclick']
    driver.execute_script(code)
    mall_link.append(driver.current_url)
    try:
        image_link.append(driver.find_element(By.XPATH, "//img[@alt='banner']").get_attribute("src"))
    except:
        image_link.append(None)
    try:
        mall_contact_no.append(driver.find_element(By.XPATH, "//*[@class='contact-info']/ul//*[contains(text(),'Landline') or contains(text(),'Contact no.')]").text)
    except:
        mall_contact_no.append(None)
    try:
        fb_link.append(driver.find_element(By.XPATH, "//*[@class='contact-info']/ul//*[contains(text(),'Facebook')]").text)
    except:
        fb_link.append(None)

    #print("contact 0", contact.text)
    #print("contact 1", landline.text)
    #print("contact 2", facebook.text)

    driver.back()
    time.sleep(2)
    
```

```python
robinsons_malls_df = pd.DataFrame({'mall_name': mall_name,
                        'mall_address': mall_address,
                        'mall_contact_no': mall_contact_no,
                        'image_link': image_link,
                        'mall_link': mall_link,
                        'fb_link': fb_link})
```

```python
robinsons_malls_df['mall_contact_no'] = robinsons_malls_df['mall_contact_no'].apply(lambda x: x.replace('Landline', "").replace("Contact no.", "").replace(":", "").strip() if x else None)
```
